Q1. category(자동차 종류)가 'compact', 'subcompact', 'suv'인 자동차의 cty(도시 연비)가
        어떻게 다른지 비교해보려고 합니다. 세 차종의 cty를 나타낸 상자 그림을 만들어보세요.

```python
df = mpg.query('category in ["compact","subcompact","suv"]')
sns.boxplot(data = df, x = 'category', y = 'cty', hue = 'category')
```

```python
order = df.groupby('category')['cty'].median().sort_values(ascending = False)
order
sns.boxplot(data=df, x='category',y='cty', order=order.index, hue='category')
```

mpg 데이터를 이용해서 분석 문제를 해결해 보세요.

Q1. category(자동차 종류)가 'compact', 'subcompact', 'suv'인 자동차의 cty(도시 연비)가
        어떻게 다른지 비교해보려고 합니다. 세 차종의 cty를 나타낸 상자 그림을 만들어보세요.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# mpg 데이터셋 불러오기
mpg = pd.read_csv('mpg.csv')

# 1. 'compact', 'subcompact', 'suv' 카테고리만 선택
df = mpg.query("category in ['compact', 'subcompact', 'suv']")

# 2. 상자 그림 그리기
sns.boxplot(x='category', y='cty', data=df)
```

Q2. 상자 그림을 중앙값이 큰 순으로 정렬하세요.

```python
# 1. category 컬럼을 그룹바이로 묶고 median() 함수로 정렬
order = df.groupby('category')['cty'].median().sort_values(ascending=False).index

# 2. 상자 그림 그리기
sns.boxplot(x='category', y='cty', data=df, order=order)
```
