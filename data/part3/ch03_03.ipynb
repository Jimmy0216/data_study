Q. 'audi'에서 생산한 자동차 중에 어떤 자동차 모델의 hwy 가 높은지 알아보려고 합니다.
    'audi'에서 생산한 자동차 중 hwy 1~5위에 해당하는 자동차의 데이터를 출력하세요.

```python
# mpg 데이터 로드
data = pd.read_csv('mpg.csv')

# 'audi'에서 생산한 자동차 중 hwy 상위 5개 추출 (mpg.query와 메소드 체이닝 사용)
top_5_audi = (data.query("manufacturer == 'audi'") \
              .sort_values(by='hwy', ascending=False) \
              .head(5))

top_5_audi
```
