Q1. mpg.csv를 불러와 복사본을 만드세요.

```python
import pandas as pd

# mpg 데이터 불러오기
mpg = pd.read_csv('mpg.csv')

# mpg 복사본 만들기
mpg_copy = mpg.copy()
mpg_copy
```

Q2. 복사본 데이터를 이용해서 cty는 city로, hwy는 highway로 변수명을 수정하세요.

```python
# 'cty'를 'city'로, 'hwy'를 'highway'로 변수명 변경
mpg_copy = mpg_copy.rename(columns={'cty': 'city', 'hwy': 'highway'})
mpg_copy
```
